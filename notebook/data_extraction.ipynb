```python
import yfinance as yf
import pandas as pd

# Define tickers and date range
tickers = ['TSLA', 'BND', 'SPY']
start_date = '2015-07-01'
end_date = '2025-07-31'

# Download data
data = yf.download(tickers, start=start_date, end=end_date, group_by='ticker', auto_adjust=False)

# Flatten column MultiIndex for clarity
data.columns = ['_'.join(col).strip() for col in data.columns.values]

# Create merged DataFrame starting with Date index
merged_df = pd.DataFrame(index=data.index)

# Process each asset
for ticker in tickers:
    # Keep OHLCV
    merged_df[f'{ticker}_Open'] = data[f'{ticker}_Open']
    merged_df[f'{ticker}_High'] = data[f'{ticker}_High']
    merged_df[f'{ticker}_Low'] = data[f'{ticker}_Low']
    merged_df[f'{ticker}_Close'] = data[f'{ticker}_Close']
    merged_df[f'{ticker}_AdjClose'] = data[f'{ticker}_Adj Close']
    merged_df[f'{ticker}_Volume'] = data[f'{ticker}_Volume']
    
    # Daily returns
    merged_df[f'{ticker}_Returns'] = merged_df[f'{ticker}_AdjClose'].pct_change()
    
    # 21-day rolling volatility (annualized)
    merged_df[f'{ticker}_Volatility'] = (
        merged_df[f'{ticker}_Returns']
        .rolling(window=21)
        .std() * (252 ** 0.5)
    )

# Drop initial NaN rows from returns/volatility calculations
merged_df = merged_df.dropna()

# Save merged dataset
merged_df.to_csv('../data/financial_data_merged.csv')

print("✅ Merged time-series dataset saved to financial_data_merged.csv")
print(merged_df.head())
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller

# =========================
# 1. Fetch & Merge Data
# =========================
tickers = ['TSLA', 'BND', 'SPY']
start_date = '2015-07-01'
end_date = '2025-07-31'

print("📥 Downloading data...")
data = yf.download(tickers, start=start_date, end=end_date, group_by='ticker', auto_adjust=False)
data.columns = ['_'.join(col).strip() for col in data.columns.values]

merged_df = pd.DataFrame(index=data.index)

for ticker in tickers:
    merged_df[f'{ticker}_Open'] = data[f'{ticker}_Open']
    merged_df[f'{ticker}_High'] = data[f'{ticker}_High']
    merged_df[f'{ticker}_Low'] = data[f'{ticker}_Low']
    merged_df[f'{ticker}_Close'] = data[f'{ticker}_Close']
    merged_df[f'{ticker}_AdjClose'] = data[f'{ticker}_Adj Close']
    merged_df[f'{ticker}_Volume'] = data[f'{ticker}_Volume']
    merged_df[f'{ticker}_Returns'] = merged_df[f'{ticker}_AdjClose'].pct_change()
    merged_df[f'{ticker}_Volatility'] = merged_df[f'{ticker}_Returns'].rolling(window=21).std() * (252 ** 0.5)

merged_df = merged_df.dropna()
merged_df.to_csv('financial_data_merged.csv')
print("✅ Merged dataset saved to financial_data_merged.csv")

# =========================
# 2. EDA Visualizations
# =========================
plt.style.use('seaborn-v0_8')

# Closing Price Trends
plt.figure(figsize=(12, 6))
for ticker in tickers:
    plt.plot(merged_df.index, merged_df[f'{ticker}_Close'], label=ticker)
plt.title("Closing Prices Over Time (2015-2025)")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.legend()
plt.show()

# Rolling Volatility
plt.figure(figsize=(12, 6))
for ticker in tickers:
    plt.plot(merged_df.index, merged_df[f'{ticker}_Volatility'], label=ticker)
plt.title("21-Day Rolling Annualized Volatility")
plt.xlabel("Date")
plt.ylabel("Volatility")
plt.legend()
plt.show()

# Daily Returns Distribution
plt.figure(figsize=(12, 6))
for ticker in tickers:
    merged_df[f'{ticker}_Returns'].hist(alpha=0.5, bins=100, label=ticker)
plt.title("Distribution of Daily Returns")
plt.xlabel("Daily Return")
plt.ylabel("Frequency")
plt.legend()
plt.show()

# =========================
# 3. Augmented Dickey-Fuller Tests
# =========================
def adf_test(series, title=''):
    """Perform Augmented Dickey-Fuller test and print results."""
    result = adfuller(series.dropna(), autolag='AIC')
    labels = ['ADF Statistic', 'p-value', '# Lags Used', 'Number of Observations Used']
    print(f'\n📊 ADF Test for {title}')
    for value, label in zip(result[0:4], labels):
        print(f'{label} : {value}')
    if result[1] <= 0.05:
        print("✅ Reject H0 → Series is Stationary")
    else:
        print("❌ Fail to Reject H0 → Series is Non-Stationary")

# Test for each asset's Close & Returns
for ticker in tickers:
    adf_test(merged_df[f'{ticker}_Close'], title=f"{ticker} Closing Prices")
    adf_test(merged_df[f'{ticker}_Returns'], title=f"{ticker} Returns")
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller

# =========================
# 1. Fetch & Merge Data
# =========================
tickers = ['TSLA', 'BND', 'SPY']
start_date = '2015-07-01'
end_date = '2025-07-31'

print("📥 Downloading data...")
data = yf.download(tickers, start=start_date, end=end_date, group_by='ticker', auto_adjust=False)
data.columns = ['_'.join(col).strip() for col in data.columns.values]

merged_df = pd.DataFrame(index=data.index)

for ticker in tickers:
    merged_df[f'{ticker}_Open'] = data[f'{ticker}_Open']
    merged_df[f'{ticker}_High'] = data[f'{ticker}_High']
    merged_df[f'{ticker}_Low'] = data[f'{ticker}_Low']
    merged_df[f'{ticker}_Close'] = data[f'{ticker}_Close']
    merged_df[f'{ticker}_AdjClose'] = data[f'{ticker}_Adj Close']
    merged_df[f'{ticker}_Volume'] = data[f'{ticker}_Volume']
    merged_df[f'{ticker}_Returns'] = merged_df[f'{ticker}_AdjClose'].pct_change()
    merged_df[f'{ticker}_Volatility'] = merged_df[f'{ticker}_Returns'].rolling(window=21).std() * (252 ** 0.5)

merged_df = merged_df.dropna()
merged_df.to_csv('../data/financial_data_merged.csv')
print("✅ Original merged dataset saved to financial_data_merged.csv")

# =========================
# 2. EDA Visualizations
# =========================
plt.style.use('seaborn-v0_8')

# Closing Price Trends
plt.figure(figsize=(12, 6))
for ticker in tickers:
    plt.plot(merged_df.index, merged_df[f'{ticker}_Close'], label=ticker)
plt.title("Closing Prices Over Time (2015-2025)")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.legend()
plt.show()

# Rolling Volatility
plt.figure(figsize=(12, 6))
for ticker in tickers:
    plt.plot(merged_df.index, merged_df[f'{ticker}_Volatility'], label=ticker)
plt.title("21-Day Rolling Annualized Volatility")
plt.xlabel("Date")
plt.ylabel("Volatility")
plt.legend()
plt.show()

# Daily Returns Distribution
plt.figure(figsize=(12, 6))
for ticker in tickers:
    merged_df[f'{ticker}_Returns'].hist(alpha=0.5, bins=100, label=ticker)
plt.title("Distribution of Daily Returns")
plt.xlabel("Daily Return")
plt.ylabel("Frequency")
plt.legend()
plt.show()

# =========================
# 3. ADF Tests + Auto-Differencing
# =========================
def adf_test(series):
    """Return ADF p-value."""
    result = adfuller(series.dropna(), autolag='AIC')
    return result[1]  # p-value

stationary_df = merged_df.copy()

for ticker in tickers:
    for col_type in ['Close', 'Returns']:
        col_name = f"{ticker}_{col_type}"
        series = stationary_df[col_name].dropna()

        # Auto-difference until p-value <= 0.05
        diff_count = 0
        p_value = adf_test(series)

        while p_value > 0.05 and len(series) > 1:
            diff_count += 1
            series = series.diff().dropna()
            p_value = adf_test(series)

        if diff_count > 0:
            print(f"🔄 {col_name} differenced {diff_count} time(s) to achieve stationarity (p={p_value:.4f})")
            # Replace with differenced series in stationary_df
            stationary_df[col_name] = stationary_df[col_name].diff(periods=diff_count)

        else:
            print(f"✅ {col_name} already stationary (p={p_value:.4f})")

# Drop NaNs from differencing
stationary_df = stationary_df.dropna()
stationary_df.to_csv('../data/financial_data_stationary.csv')
print("✅ Stationary dataset saved to financial_data_stationary.csv")
```
